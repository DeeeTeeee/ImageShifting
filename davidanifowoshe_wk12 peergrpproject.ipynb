## Define a function to shift the image up, down, left and right using _scipy.ndimage.interpolation_  by a certain number of pixels :

Shifting Images

Hint: To shift images we can use the shift function from scipy.ndimage.interpolation

**2.1 Shift Image Up**

Pick an image from the dataset and view it.

Now write a function to shift the image up by 1 pixel (You can test your function by shifting the image up by about 5 pixels and viewing it to observe the change in the digits position)
### Shift Image Up

```python
#using the shift function from scipy.ndimage.interpolation 
from scipy.ndimage.interpolation import shift
```

The shift function takes three arguments: the input image, the shift amount as a tuple of (-num_pixels, 0) to shift the image up, and cval=0 to fill any empty spaces with 0.

lets call the shift_image_up function and pass in the image and the desired number of pixels to shift the image up:

```python
#lets call the shift_image_up function:
def shift_image_up(image, num_pixels):
    shifted_image = shift(image, (-num_pixels, 0), cval=0)
    return shifted_image
```

```python
# pass in the image and the desired number of pixels to shift the image up:
shifted_imageup = shift_image_up(image, 5)
```

```python
#Display the shifted image using plt.imshow:
plt.imshow(shifted_imageup);
```


### 2.2 Shift Image Down 

**Shift Image Down**

Pick an image from the dataset and view it.

Now write a function to shift the image down by 1 pixel (You can test your function by shifting the image down by about 5 pixels and viewing it to observe the change in the digits position)

The shift_image_down function takes two arguments: the input image and the number of pixels to shift the image down by. The shift function is called with the shift amount as a tuple of (num_pixels, 0) to shift the image down, and cval=0 to fill any empty spaces with 0.

```python
#lets call the shift_image_up function:
def shift_image_down(image, num_pixels):
    shifted_imagedw = shift(image, (num_pixels, 0), cval=0)
    return shifted_imagedw
```

To test the function, you can load an image and display it, then call the shift_image_down function with the desired number of pixels to shift the image down by, and display the resulting image. For example, to shift the image down by 5 pixels, you can do:

```python
# Shift the image down by 5 pixels
shifted_image = shift_image_down(image, 5)

# Display the shifted image
plt.imshow(shifted_image);
```

### 2.3 Shift Image Right 

**2.3 Shift Image Right**

Pick an image from the dataset and view it.

Now write a function to shift the image right by 1 pixel (You can test your function by shifting the image to the right by about 5 pixels and viewing it to observe the change in the digits position)

The shift_image_right function takes two arguments: the input image and the number of pixels to shift the image right by. The shift function is called with the shift amount as a tuple of (0, num_pixels) to shift the image right, and cval=0 to fill any empty spaces with 0.

```python
def shift_image_right(image, num_pixels):
    shifted_image = shift(image, (0, num_pixels), cval=0)
    return shifted_image
```

To test the function, you can load an image and display it, then call the shift_image_right function with the desired number of pixels to shift the image right by, and display the resulting image. For example, to shift the image right by 5 pixels, you can do:

```python
# Shift the image right by 5 pixels
shifted_image = shift_image_right(image, 5)

# Display the shifted image
plt.imshow(shifted_image);
```

You can adjust the number of pixels to shift the image right by changing the second argument of the shift_image_right function call.

### 2.4 Shift Image Left 

**2.4 Shift Image Left**

Pick an image from the dataset and view it.

Now write a function to shift the image left by 1 pixel (You can test your function by shifting the image to the left by about 5 pixels and viewing it to observe the change in the digits position)

The shift_image_left function takes two arguments: the input image and the number of pixels to shift the image left by. The shift function is called with the shift amount as a tuple of (0, -num_pixels) to shift the image left, and cval=0 to fill any empty spaces with 0.

```python
def shift_image_left(image, num_pixels):
    shifted_image = shift(image, (0, -num_pixels), cval=0)
    return shifted_image
```

To test the function, you can load an image and display it, then call the shift_image_left function with the desired number of pixels to shift the image left by, and display the resulting image. For example, to shift the image left by 5 pixels, you can do:

```python
# Shift the image left by 5 pixels
shifted_image = shift_image_left(image, 5)

# Display the shifted image
plt.imshow(shifted_image);
```
